# Using the sklearn (scikit-learn) library

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/cypherkrish/Machine-Learning-Basics/main/Data/Advertising.csv')
```

```python
df.head()
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 3, figsize = (15, 4))

ax[0].plot(df['TV'], df['sales'], 'o')
ax[0].set_title("TV Spend")
ax[0].set_ylabel('Sales')

ax[1].plot(df['radio'], df['sales'], 'o')
ax[1].set_title('Radio Spend')
ax[1].set_ylabel('Sales')

ax[2].plot(df['newspaper'], df['sales'], 'o')
ax[2].set_title("Newspaper Spend")
ax[2].set_ylabel("Sales")

plt.tight_layout()
```

```python
sns.pairplot(data=df)
```

```python
df.head()
```

```python
X = df.drop('sales', axis = 1)
```

```python
X.head()
```

```python
y = df['sales']
```

```python
# seperate the data set to train and test data split

from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
help(train_test_split)
```

```python
len(X)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
33 * 200 / 100
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr_model = LinearRegression()
```

```python
lr_model.fit(X_train, y_train)
```

```python
test_pred = lr_model.predict(X_test)
```

```python
X_test.head()
```

```python
y_test.head()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
df['sales'].mean()
```

```python
sns.histplot(data = df, x= 'sales', bins=20)
```

```python
mean_absolute_error(y_test, test_pred)
```

```python
MSE = mean_squared_error(y_test, test_pred)
```

```python
MSE
```

```python
RMSE = np.sqrt(MSE)
```

```python
RMSE
```

```python
# Plot the residuals

test_residuals = y_test - test_pred
```

```python
test_residuals
```

```python
sns.scatterplot(x = y_test, y =test_residuals)
plt.axhline(y=0, color = 'red')
```

```python
sns.displot(test_residuals, kde = True, bins = 25)
```

```python
import scipy as sp
```

```python
fig, ax = plt.subplots(figsize = (10, 8), dpi = 100)

_ = sp.stats.probplot(test_residuals, plot = ax)
```

```python
final_model = LinearRegression()
```

```python
final_model.fit(X, y)
```

```python
final_model.coef_
```

```python
X.head()
```

```python
y_hat = final_model.predict(X)
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 3, figsize = (15, 4))

ax[0].plot(df['TV'], df['sales'], 'o')
ax[0].plot(df['TV'], y_hat, 'o', color = 'red')
ax[0].set_title("TV Spend")
ax[0].set_ylabel('Sales')

ax[1].plot(df['radio'], df['sales'], 'o')
ax[1].plot(df['radio'], y_hat, 'o', color = 'red')
ax[1].set_title('Radio Spend')
ax[1].set_ylabel('Sales')

ax[2].plot(df['newspaper'], df['sales'], 'o')
ax[2].plot(df['newspaper'], y_hat, 'o', color = 'red')
ax[2].set_title("Newspaper Spend")
ax[2].set_ylabel("Sales")

plt.tight_layout()
```
